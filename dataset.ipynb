```python
import json

def analyze_jsonl_file(file_path):
    # Read the JSONL file
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Number of rows in the dataset
    num_rows = len(lines)
    print(f"Number of rows: {num_rows}")
    
    # Number of columns in the dataset (assumes consistent structure)
    num_columns = len(json.loads(lines[0]).keys()) if num_rows > 0 else 0
    print(f"Number of columns: {num_columns} \n")
```

```python
# Call the function to analyze the JSONL file
analyze_jsonl_file('new_research_papers.jsonl')
```

```python
# Call the function to analyze the JSONL file
analyze_jsonl_file('dataset.jsonl')
```
